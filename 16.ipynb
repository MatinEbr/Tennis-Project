```python
merge_home_away = pd.merge(home_team,away_team,on = 'match_id')
rank_home_players = merge_home_away[['name_x','current_rank_x']].rename(columns={'name_x':'name','current_rank_x':'rank'})
rank_away_players = merge_home_away[['name_y','current_rank_y']].rename(columns={'name_y':'name','current_rank_y':'rank'})
rank_players = pd.concat([rank_away_players,rank_home_players])
rank_players = rank_players.groupby('name', as_index=False)['rank'].mean().dropna()
rank_players = rank_players.loc[rank_players['rank']<=10]
rank_players
```

```python
home_away_event = pd.merge(merge_home_away,event,on = 'match_id')
home_away_event.loc[home_away_event['winner_code']==1,'winner']=home_away_event['name_x']
home_away_event.loc[home_away_event['winner_code']==2,'winner']=home_away_event['name_y']
home_players = home_away_event[['match_id','name_x','winner']].rename(columns={'name_x':'player_name','match_id':'match_id','winner':'winner'})
away_players = home_away_event[['match_id','name_y','winner' ]].rename(columns={'name_y':'player_name','match_id':'match_id','winner':'winner'})
players_name = pd.concat([home_players, away_players])
players_name_won = pd.merge(players_name,rank_players,left_on='player_name',right_on='name')
players_name_won = players_name_won['winner'].value_counts().reset_index()
players_name_won
```
