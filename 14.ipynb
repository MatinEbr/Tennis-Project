```python
tournament = tournament.dropna(subset=['ground_type'])

ground_type = tournament[['match_id' , 'ground_type']]

ground_type.head()
```

```python
groupby = ground_type.groupby('ground_type')['match_id'].count().reset_index().sort_values(by = 'match_id' , ascending= False)

groupby = groupby.rename(columns={'match_id' : 'Number of games'})

groupby
```
