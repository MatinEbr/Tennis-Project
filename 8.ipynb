```python
import pandas as pd


away_team_score = pd.read_csv('E:\\daneshkar\\Tennis Schema\\output\\csv output\\away_team_score.csv')
away_team = pd.read_csv('E:\\daneshkar\\Tennis Schema\\output\\csv output\\away_team.csv')
event = pd.read_csv('E:\\daneshkar\\Tennis Schema\\output\\csv output\\event.csv')
home_team_score = pd.read_csv('E:\\daneshkar\\Tennis Schema\\output\\csv output\\home_team_score.csv')
home_team = pd.read_csv('E:\\daneshkar\\Tennis Schema\\output\\csv output\\home_team.csv')
raw_odds_parquet = pd.read_csv('E:\\daneshkar\\Tennis Schema\\output\\csv output\\raw_odds_parquet.csv')
raw_point_by_point_parquet = pd.read_csv('E:\\daneshkar\\Tennis Schema\\output\\csv output\\raw_point_by_point_parquet.csv')
raw_statistics_parquet = pd.read_csv('E:\\daneshkar\\Tennis Schema\\output\\csv output\\raw_statistics_parquet.csv')
raw_tennis_power_parquet = pd.read_csv('E:\\daneshkar\\Tennis Schema\\output\\csv output\\raw_tennis_power_parquet.csv')
raw_votes_parquet = pd.read_csv('E:\\daneshkar\\Tennis Schema\\output\\csv output\\raw_votes_parquet.csv')
round = pd.read_csv('E:\\daneshkar\\Tennis Schema\\output\\csv output\\round.csv')
season = pd.read_csv('E:\\daneshkar\\Tennis Schema\\output\\csv output\\season.csv')
time = pd.read_csv('E:\\daneshkar\\Tennis Schema\\output\\csv output\\time.csv')
tournament = pd.read_csv('E:\\daneshkar\\Tennis Schema\\output\\csv output\\tournament.csv')
venue = pd.read_csv('E:\\daneshkar\\Tennis Schema\\output\\csv output\\venue.csv')

```

```python
merged_home_away = pd.merge(home_team , away_team , how = 'left' , on ='match_id' , suffixes=('_home' , '_away'))

merged_home_away_1 = merged_home_away[['match_id' , 'full_name_home' ,'player_id_home' , 'gender_home' , 'full_name_away' ,'player_id_away', 'gender_away']]

merged_home_away_1.head()
```

```python
raw_statistics_parquet = raw_statistics_parquet[['match_id' , 'statistic_name' , 'home_stat' , 'away_stat' ,'period']]

raw_statistics_parquet_double_faults = raw_statistics_parquet[(raw_statistics_parquet['statistic_name'] == 'double_faults') & (raw_statistics_parquet['period'] == "ALL")]

raw_statistics_parquet_double_faults.head()
```

```python
merged_home_away_statistics = pd.merge(merged_home_away_1 , raw_statistics_parquet_double_faults , how = 'left' , on = 'match_id')

merged_home_away_statistics.head()
```

```python
merged_home_away_statistics = merged_home_away_statistics.dropna()


merged_home_away_statistics['home_away_stat_sum'] =merged_home_away_statistics['home_stat'].astype(int) + merged_home_away_statistics['away_stat'].astype(int)
```

```python
merged_home_away_statistics.head()
```

```python
groupby = merged_home_away_statistics.groupby(['gender_home' , 'gender_away'])['home_away_stat_sum'].sum().reset_index()

groupby.head()
```

```python
M = merged_home_away_statistics[(merged_home_away_statistics['gender_home'] == 'M') & (merged_home_away_statistics['gender_away'] == 'M')]

M_groupby= M.groupby(['gender_home' , 'gender_away'])['home_away_stat_sum'].mean().reset_index()

M = merged_home_away_statistics[(merged_home_away_statistics['gender_home'] == 'F') & (merged_home_away_statistics['gender_away'] == 'F')]

F_groupby = M.groupby(['gender_home' , 'gender_away'])['home_away_stat_sum'].mean().reset_index()
```

```python
M_groupby.rename(columns={'home_away_stat_sum': 'home_away_stat_mean '}, inplace=True)

M_groupby
```
