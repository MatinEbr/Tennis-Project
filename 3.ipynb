```python
merged_home_away = pd.merge(home_team , away_team , how = 'left' , on ='match_id' , suffixes=('_home' , '_away'))
merged_home_away = merged_home_away[['match_id' ,'full_name_home' , 'full_name_away']]

merged = pd.merge(merged_home_away, event[['match_id', 'winner_code']], how='left', on='match_id')

merged.head()
```

```python
def winner_name(row):
    if row['winner_code'] == 1:
        return row['full_name_home']
    elif row['winner_code'] == 2:
        return row['full_name_away']
    else:
        return None

merged['winner_name'] = merged.apply(winner_name, axis=1)

merged.head()
```

```python
win_counts = merged['winner_name'].value_counts().reset_index()

first_winner = win_counts.loc[0, 'winner_name'] 
win_count = win_counts.loc[0, 'count'] 

win_counts.head()
```

```python
print(f"The player with the most wins is {first_winner} with {win_count} wins.")
```
