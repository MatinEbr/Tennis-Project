```python
import pandas as pd

time = pd.read_csv('D:\\daneshkar\\Tennis Schema\\output\\csv output\\time.csv')
```

```python
time.iloc[10:21]
```

```python
time = time.fillna(0)
time['total_duration_sec'] = time[[ 'period_1', 'period_2' , 'period_3' , 'period_4' , 'period_5']].sum(axis = 1)
time['time_duration_hour'] = time['total_duration_sec'] / 3600
time = time.sort_values(by = 'time_duration_hour' , ascending= False).reset_index()

time.head()
```

```python
longest_match = time.loc[0, 'match_id'] 
time_duration = time.loc[0, 'time_duration_hour']

print(f'The match with ID {longest_match} had the longest duration, lasting {time_duration:.1f} hours.')
```
